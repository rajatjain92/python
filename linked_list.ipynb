```python
class Node:
    def __init__(self,data):
        self.item = data
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
        
    def traverse_list(self):
        if self.head is None:
            print('List has no element')
            return
        else:
            n = self.head
            while n is not None:
                print(n.item,end = " ")
                n = n.next
            print()
    
    def insert_at_start(self,data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        
    def insert_at_end(self,data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        else:
            n = self.head
            while n.next is not None:
                n = n.next
            n.next = new_node
            
            
    def insert_after_item(self,x,data):
        n = self.head
        while n is not None:
            if n.item == x:
                break
            n = n.next
        if n is None:
            print("item not present")
        else:
            new_node = Node(data)
            new_node.next = n.next
            n.next = new_node
            
    def insert_before_item(self,x,data):
        n = self.head
        if n is None:
            print("List has no element")
            return
        
        if x == n.item:
            new_node = Node(data)
            new_node.next = n
            self.head = new_node
            return
        
        while n.next is not None:
            if n.next.item == x:
                break
            n = n.next
        if n.next is None:
            print("element is not present")
        else:
            new_node = Node(data)
            new_node.next = n.next
            n.next = new_node
            
    
    def insert_at_index(self,index,data):
        if index == 0:
            self.insert_at_start(self,data)
        
        n = self.head
        i = 0
        while i < index-1 and n is not None:
            n = n.next
            i = i+1
        
        if n is None:
            print('out of index')
        else:
            new_node = Node(data)
            new_node.next = n.next
            n.next = new_node
            
            
    def get_count(self):
        if self.head is None:
            return 0
        n = self.head
        count = 0
        while n is not None:
            count = count + 1
            n = n.next
        print(count)
        
    
    def search(self,x):
        if self.head is None:
            print("list has no elements")
            return
        n = self.head
        i = 0
        while n is not None:
            if n.item == x:
                print("item found at:",i)
                return True
            n = n.next
            i = i+1
        print("Item not found")
        return False
                
    
    def make_new_list(self):
        nums = int(input("number of nodes:"))
        if nums == 0:
            return
        for i in range(nums):
            value = int(input("enter value of node:"))
            self.insert_at_end(value)
            
    
    def delete_at_start(self):
        if self.head is None:
            print("the list is empty")
            return
        self.head = self.head.next
        
        
    def delete_at_end(self):
        if self.head is None:
            print("the list is empty")
            return
        n = self.head
        while n.next.next is not None:
            n = n.next
        n.next = None
        
    
    def delete_by_value(self,x):
        if self.head is None:
            print("No element")
            return
        
        #check first element
        if self.head.item == x:
            self.head = self.head.next
            return
        
        #check from 2 onwards
        n = self.head
        while n.next is not None:
            if n.next.item == x:
                break
            n = n.next
        
        if n.next is None:
            print('Item not found')
        else:
            n.next = n.next.next
            
    
    def reverse(self):
        prev = None
        n = self.head
        while n is not None:
            next = n.next
            n.next = prev
            prev = n
            n = next
        self.head = prev
            
            
    
    def bubble_sort_value(self):
        #end represents end of each pass
        #gradually end will be decreasing
        end = None
        while end != self.head:
            p = self.head
            while p.next != end:        # using p.next != end bcoz when p is last element we do not want to run go inside while loop 
                q = p.next
                if p.item > q.item:
                    p.item,q.item = q.item,p.item
                p = p.next
            end = p
            
            
    #its same as interchanging 2 linked list node in pairs using p,q,r pointers
    def bubble_sort_link(self):
        end = None
        while end != self.head:
            r = p = self.head
            while p.next != end:
                q = p.next
                if p.item > q.item:
                    p.next = q.next
                    q.next = p
                    if p != self.head:
                        r.next = q
                    else:
                        self.head = q
                    p,q = q,p
                r = p
                p = p.next
            end = p
                
```

```python
if __name__ == '__main__':
    llist = LinkedList()
    llist.head = Node(1)
    second = Node(2)
    third = Node(3)
    
    llist.head.next = second
    second.next = third
    
    llist.traverse_list()
    llist.insert_at_start(0)
    llist.traverse_list()
    llist.insert_at_end(4)
    llist.traverse_list()
    llist.insert_after_item(3,5)
    llist.traverse_list()
    llist.insert_before_item(3,9)
    llist.traverse_list()
    llist.insert_at_index(3,777)
    llist.traverse_list()
    llist.get_count()
    llist.search(3)
    
    llist2 = LinkedList()
    llist2.make_new_list()
    
    llist.traverse_list()
    llist.delete_at_start()
    llist.traverse_list()
    llist.delete_at_end()
    llist.traverse_list()
```
