```python
class Node:
    def __init__(self,data):
        self.item = data
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
        
    def traverse_list(self):
        if self.head is None:
            print('List has no element')
            return
        else:
            n = self.head
            while n is not None:
                print(n.item,end = " ")
                n = n.next
            print()
    
    def insert_at_start(self,data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        
    def insert_at_end(self,data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        else:
            n = self.head
            while n.next is not None:
                n = n.next
            n.next = new_node
            
            
    def insert_after_item(self,x,data):
        n = self.head
        while n is not None:
            if n.item == x:
                break
            n = n.next
        if n is None:
            print("item not present")
        else:
            new_node = Node(data)
            new_node.next = n.next
            n.next = new_node
            
    def insert_before_item(self,x,data):
        n = self.head
        if n is None:
            print("List has no element")
            return
        
        if x == n.item:
            new_node = Node(data)
            new_node.next = n
            self.head = new_node
            return
        
        while n.next is not None:
            if n.next.item == x:
                break
            n = n.next
        if n.next is None:
            print("element is not present")
        else:
            new_node = Node(data)
            new_node.next = n.next
            n.next = new_node
            
    
    def insert_at_index(self,index,data):
        if index == 0:
            self.insert_at_start(self,data)
        
        n = self.head
        i = 0
        while i < index-1 and n is not None:
            n = n.next
            i = i+1
        
        if n is None:
            print('out of index')
        else:
            new_node = Node(data)
            new_node.next = n.next
            n.next = new_node
            
            
    def get_count(self):
        if self.head is None:
            return 0
        n = self.head
        count = 0
        while n is not None:
            count = count + 1
            n = n.next
        print(count)
        
    
    def search(self,x):
        if self.head is None:
            print("list has no elements")
            return
        n = self.head
        i = 0
        while n is not None:
            if n.item == x:
                print("item found at:",i)
                return True
            n = n.next
            i = i+1
        print("Item not found")
        return False
                
    
    def make_new_list(self):
        nums = int(input("number of nodes:"))
        if nums == 0:
            return
        for i in range(nums):
            value = int(input("enter value of node:"))
            self.insert_at_end(value)
            
    
    def delete_at_start(self):
        if self.head is None:
            print("the list is empty")
            return
        self.head = self.head.next
        
        
    def delete_at_end(self):
        if self.head is None:
            print("the list is empty")
            return
        n = self.head
        while n.next.next is not None:
            n = n.next
        n.next = None
        
    
    def delete_by_value(self,x):
        if self.head is None:
            print("No element")
            return
        
        #check first element
        if self.head.item == x:
            self.head = self.head.next
            return
        
        #check from 2 onwards
        n = self.head
        while n.next is not None:
            if n.next.item == x:
                break
            n = n.next
        
        if n.next is None:
            print('Item not found')
        else:
            n.next = n.next.next
            
    
    def reverse(self):
        prev = None
        n = self.head
        while n is not None:
            next = n.next
            n.next = prev
            prev = n
            n = next
        self.head = prev
            
            
    
    def bubble_sort_value(self):
        #end represents end of each pass
        #gradually end will be decreasing
        end = None
        while end != self.head:
            p = self.head
            while p.next != end:        # using p.next != end bcoz when p is last element we do not want to run go inside while loop 
                q = p.next
                if p.item > q.item:
                    p.item,q.item = q.item,p.item
                p = p.next
            end = p
            
            
    #its same as interchanging 2 linked list node in pairs using p,q,r pointers
    def bubble_sort_link(self):
        end = None
        while end != self.head:
            r = p = self.head
            while p.next != end:
                q = p.next
                if p.item > q.item:
                    p.next = q.next
                    q.next = p
                    if p != self.head:
                        r.next = q
                    else:
                        self.head = q
                    p,q = q,p
                r = p
                p = p.next
            end = p
      
    
    def merge_helper(self, list2):
        merged_list = LinkedList()
        merged_list.head = self.merge_by_newlist(self.head, list2.head)
        return merged_list

    def merge_by_newlist(self, p, q):
        if p.item <= q.item:
            startNode = Node(p.item)
            p = p.next
        else:
            startNode = Node(q.item)
            q = q.next

        em = startNode

        while p is not None and q is not None:
            if p.item <= q.item:
                em.next = Node(p.item)
                p = p.next
            else:
                em.next = Node(q.item)
                q = q.next
            em = em.next

        while p is not None:
            em.next = Node(p.item)
            p = p.next
            em = em.next

        while q is not None:
            em.next = Node(q.item)
            q = q.next
            em = em.next

        return startNode
    
    
    
    
    
    def detectAndRemoveLoop(self): 
        slow_p = fast_p = self.head 
        while(slow_p and fast_p and fast_p.next): 
            slow_p = slow_p.next 
            fast_p = fast_p.next.next
          
            # If slow_p and fast_p meet at some poin 
            # then there is a loop 
            if slow_p == fast_p: 
                self.removeLoop(slow_p) 
                  
                # Return 1 to indicate that loop if found 
                return 1 
  
        # Return 0 to indicate that there is no loop 
        return 0
  
    # Function to remove loop 
    # loop node-> Pointer to one of the loop nodes 
    # head --> Pointer to the start node of the 
    # linked list 
    def removeLoop(self, loop_node): 
          
        # Set a pointer to the beginning of the linked  
        # list and move it one by one to find the first 
        # node which is part of the linked list 
        ptr1 = self.head 
        while(1): 
            # Now start a pointer from loop_node and check 
            # if it ever reaches ptr2 
            ptr2 = loop_node 
            while(ptr2.next != loop_node and ptr2.next != ptr1): 
                ptr2 = ptr2.next
              
            # If ptr2 reached ptr1 then there is a loop. 
            # So break the loop 
            if ptr2.next == ptr1 :  
                break 
              
            ptr1 = ptr1.next
          
        # After the end of loop ptr2 is the lsat node of  
        # the loop. So make next of ptr2 as NULL 
        ptr2.next = None
```
